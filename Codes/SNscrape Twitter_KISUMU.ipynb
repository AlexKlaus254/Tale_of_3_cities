```python
#pip install git+https://github.com/JustAnotherArchivist/snscrape.git
```

```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
#Tweets from Kisumu
scraper = sntwitter.TwitterSearchScraper('🤣 geocode:-0.091702,34.768675,1000km since:2022-04-01 until:2023-05-01')
```

```python
tweet_data=[]
for i,tweet in enumerate(scraper.get_items()): 
    tweet_list=[tweet.date,
            tweet.id, 
            tweet.content,
            tweet.user.username,
            tweet.likeCount,
            tweet.retweetCount]
    tweet_data.append(tweet_list)
    if i==50000 :
        break
    
```

```python
tweet_data
```

```python
tweet_df = pd.DataFrame(tweet_data, columns=['date', 'id', 'content', 'username', 'like', 'retweet'])
```

```python
tweet_df
```

```python
from datetime import datetime
```

```python
# Convert the 'Date' column to timezone-unaware datetime so as to save in excel
tweet_df['date'] = tweet_df['date'].dt.tz_convert(None)
```
